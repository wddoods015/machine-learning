# iris

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# iris dataset 로드하기
iris = load_iris()

iris.keys()

# DESCR는 해당 데이터셋의 데이터 정보

print(iris['target_names'])
```

```python
# data를 df로 변환하여 'label'컬럼을 만들어 traget으로 지정하고 확인.
import pandas as pd 
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df.head(10)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

%matplotlib inline

# boxplot으로 확인, orient는 세로로 볼지 가로로 볼지 확인하는
# length는 겹치는 구간이 많아 해당 조건으로는 불류가 쉽지 않음
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'sepal length (cm)', y = 'species', data = iris_df, orient = 'h')
```

```python
# width는 length보다 겹치는 구간이 심해 큰 영향을 주긴 어려워 보임
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'sepal width (cm)', y = 'species', data = iris_df, orient = 'h')
```

```python
# petal로는 0번인 세토사를 분류하기 쉽지만 다른 두 종에 대해서는 분류는 까다롭다.
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'petal length (cm)', y = 'species', data = iris_df, orient = 'h')
```

```python
# 시각화를 통해 붓꽃의 품종별로 분포가 분리되었음을 알 수 있다.
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(iris_df, hue = 'species' )
```

```python
# 그래프 중에서 가장 분류에 영향이 큰 것 같은 두가지 컬럼을 중심으로 그래프를 작성한다.
sns.pairplot(data = iris_df,
             vars = ['petal length (cm)', 'petal width (cm)'],
             hue = 'species',
             height = 5
            )
```
